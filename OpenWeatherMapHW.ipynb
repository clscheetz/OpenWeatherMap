```python
import json
import requests
from config import api_key
import matplotlib.pyplot as plt
import pandas as pd
import random 
from citipy import citipy
import time
```

```python
base_url = f"http://api.openweathermap.org/data/2.5/weather?APPID={api_key}&q="
```

```python
results = []

city_processed_count = 0

while city_processed_count < 500:
    
    random_lat = random.uniform(-90,90)
    random_long = random.uniform(-180,180)
    
    city = citipy.nearest_city(random_lat, random_long)
    # How to get city name & country code:
    #city.city_name
    #city.country_code
    
    query_url = f"{base_url}{city.city_name},{city.country_code}"
    
    # temporarily commented out to preserve API requests
    print(f"Requesting weather data at {query_url}")
    response = requests.get(query_url).json()
    
    if int(response["cod"]) == 200:
        result = {}
        
        # Get city
        result["City"] = response["name"]
        
        # Get country
        result["Country"] = response["sys"]["country"]
                      
        # Get latitude 
        result["Lat"] = response["coord"]["lat"]
                                  
        # Get long
        result["Long"] = response["coord"]["lon"]
    
        # Get temperature(F)
        result["Temp"] = round((float(response["main"]["temp"]) - 273) * 1.8 + 32, 1)
    
        # Get humidity(%)
        result["Humidity"] = response["main"]["humidity"]
    
        # Get cloudiness(%)
        result["Clouds"] = response["clouds"]["all"]
    
        # Get wind speed(mph)
        result["Wind Speed"] = response["wind"]["speed"]
                
        results.append(result)
        
        city_processed_count += 1
        
        print(f"Processed City Name: {response['name']}, {response['sys']['country']}. City Number: {response['id']}")
              
    elif int(response["cod"]) == 404:
        print(f"{response['message']}: {city.city_name}, {city.country_code}. Attempting new city lookup.")
              
    # API only allows 60 calls/min
    time.sleep(1)
```

```python
weather_py = pd.DataFrame(results)
weather_py.head()
```

```python
weather_py.to_csv("weather_data.csv", sep=',', index=False)
```
